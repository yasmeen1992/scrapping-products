```python
r=requests.get('https://www.chinasichuanfood.com/recipe-index/')
soup=BeautifulSoup(r.content,'lxml')
```

```python
categorylist=soup.find_all('option',class_='level-0')
print(categorylist)
```

```python
categoryNames=[]
categoryNum=[]
for category in categorylist:
    categoryNames.append(category.text.strip().replace('\xa0\xa0','').split('(')[0].replace(' ','-'))
    categoryNum.append(category.text.strip().replace('\xa0\xa0','').split('(')[1].replace(')',''))
print(categoryNames)
print(categoryNum)
```

```python
links=[]
for name in categoryNames:
    link='https://www.chinasichuanfood.com/'+name+'/'
    links.append(link)
links
```

```python
testedLink='https://www.chinasichuanfood.com/beef-and-lamb/'
r=requests.get(testedLink)
soup=BeautifulSoup(r.content,'lxml')
```

```python
recipeslist = soup.find_all("ul",{"class":"feast-category-index-list fsri-list feast-grid-half feast-desktop-grid-fourth"}) 
```

```python
link_recipe=[]
name_recipe=[]
for recipe in recipeslist:
    #print(recipe)
    anchor_tags = recipe.select("a") 
    for i in anchor_tags : 
        link_URL = i.get('href') 
        link_recipe.append(link_URL)
        name_recipe.append(i.text.strip())
print(link_recipe)
print(len(link_recipe))
print(name_recipe)
```

```python
date_posts_list=[]
number_comments_list=[]
number_shares_list=[]
for link in link_recipe:
    r=requests.get(link)
    soup2=BeautifulSoup(r.content,'lxml')
    date_posts_list.append(soup2.find("time",class_="entry-modified-time").text)
    number_comments_list.append(soup2.find("span",class_="entry-comments-link").text.split(' ')[0])
    item_share=soup2.find("span",class_="dpsp-total-share-count").text
    if item_share.find('K') != -1:
       item_share=int(float(item_share.replace('K',''))*1000)
    else:
        item_share=int(item_share)
    number_shares_list.append(item_share)
print(date_posts_list)
print(number_comments_list)
print(number_shares_list)
```

```python
df2['date of post'] = date_posts_list
df2['commentsNumbers'] = number_comments_list
df2['sharesNumbers'] = number_shares_list

print(display(df2))
```

```python
images2 = content_page.find_all('p') 
 
all_pic = [] 
 
for c in images2: 
         
        a = c.select("a") 
        for i in a : 
            link_URL = i.get('href') 
            if link_URL.startswith("https://www.chinasichuanfood.com/wp-content/uploads") == True : 
                print(link_URL) 
                all_pic.append(link_URL) 
             
             
 
```

```python
import shutil
counter = 1 
for v in all_pic: 
    r = requests.get(v, stream=True) 
    if r.status_code == 200: 
        with open("beef-and-lamb\\beef-and-lamb" + "_" + str(counter) + ".png", 'wb') as f: 
            r.raw.decode_content = True 
            shutil.copyfileobj(r.raw, f) 
    counter+=1
```

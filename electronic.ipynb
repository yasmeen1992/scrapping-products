```python
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys 
from bs4 import BeautifulSoup 
from time import sleep 
import csv 
import xlsxwriter 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
import pandas as pd 
from selenium.webdriver.support import expected_conditions as EC 
import os 
import wget
import requests
```

```python
r=requests.get('https://latop10.it/')
soup=BeautifulSoup(r.content,'lxml')
```

```python
category1_list=[]
category2_list=[]
category_title=[]
category_url=[]
ul_menu=soup.select_one('ul[id=mega-menu-menu-1]')
for main_category in ul_menu:
     anchor_tags = main_category.select_one("a").text
     #print(anchor_tags)
     if anchor_tags=='Elettronica':
       # category1_list.append(anchor_tags)
        #print(anchor_tags)
        sub_categories=main_category.select("ul[class='mega-sub-menu']  li[class='mega-menu-column mega-menu-columns-3-of-12'] ul li ul a")
        #print(sub_categories)
        #category1_list=[ anchor_tags for i in range(0,len(sub_categories))]
        #category1_list=[]
        #category2_list=[]
        for sub_category in sub_categories:
            #print(sub_category.text)
            #category2_list.append(sub_category.text)
            #print(sub_category.get('href') )
            r2=requests.get(sub_category.get('href'))
            soup2=BeautifulSoup(r2.content,'lxml')
            sub_categories_link=soup2.find_all("a",class_="entry-title-link")
            category1_list+=[ anchor_tags for i in range(0,len(sub_categories_link))]
            category2_list+=[ sub_category.text for i in range(0,len(sub_categories_link))]
            for sub_category in sub_categories_link:
                #print(sub_category.text)
                category_title.append(sub_category.text)
                #print(sub_category.get('href') )
                category_url.append(sub_category.get('href'))
     



```

```python
df= pd.DataFrame( 
    {
    'Title': category_title, 
     'URL': category_url, 
     'Category 1':category1_list,
     'Category 2':category2_list
    }) 
print(display(df))
```

```python
df.to_csv('prezzi_electronic.csv', index=False,encoding='utf-8-sig')
```
